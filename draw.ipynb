```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
blank = np.zeros((500, 500, 3), dtype='uint8')
cv.imshow("Blank", blank)
cv.waitKey(0)
```

```python
# blank[:] = 0,0,255
# cv.imshow('Red', blank)
# cv.waitKey(0)
```

```python
cv.rectangle(blank, (0,0), (250,500), (132,177,101), thickness=8)
cv.imshow("Rectangle", blank)
cv.waitKey(0)
```

```python
cv.circle(blank, (blank.shape[1]//2, blank.shape[0]//2), 40, (239,177,101), thickness=1)
cv.imshow('Circle', blank)
cv.waitKey(0)
```

```python
img = cv.imread('Photos/cats 2.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# cv.imshow('Gray', gray)
# cv.waitKey(0)
```

```python
blur = cv.GaussianBlur(img, (7,7), cv.BORDER_DEFAULT)
cv.imshow('Blur', blur)
cv.waitKey(0)
```

```python
canny = cv.Canny(blur, 125, 175)
cv.imshow('Edges', canny)
cv.waitKey(0)
```

```python
img = cv.imread('Photos/park.jpg')
cv.imshow('Boston', img)
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
cv.imshow('HSV', hsv)

cv.waitKey(0)
```

```python
b,g,r = cv.split(img)
cv.imshow('Blue', b)
cv.imshow('Red', r)
cv.imshow('Green', g)

cv.waitKey(0)
```

```python
gray_hist = cv.calcHist([gray], [0], None, [256], [0,256])
plt.figure()
plt.xlabel('Bins')
plt.ylabel('No of pixels')
plt.plot(gray_hist)
plt.xlim([0,256])
plt.show()
```

```python
colors = ('b', 'g', 'r')
for i, col in enumerate(colors):
    hist = cv.calcHist([img], [i], None, [256], [0,256])
    plt.plot(hist, color=col)
    plt.xlim([0,256])
plt.xlabel('Intensity')
plt.ylabel('No. of Pixels')
plt.show()
```
